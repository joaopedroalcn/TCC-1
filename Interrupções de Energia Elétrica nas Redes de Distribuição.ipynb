Cada ponto na curva representa um par sensibilidade/especificidade correspondente a um limiar específico, permitindo a visualização do trade-off entre detectar verdadeiros positivos e evitar falsos positivos. A área sob a curva (AUC) quantifica a capacidade do modelo de distinguir entre classes, sendo que um valor de AUC próximo a 1 indica um bom desempenho, enquanto um valor de 0,5 indica um desempenho equivalente ao de uma classificação aleatória.

<h2>Desenvolvimento</h2>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o arquivo CSV
file_path = 'C:/Users/u11886/Desktop/TCC 1/interrupcoes-energia-eletrica-2024.csv'
columns_needed = ['NumOrdemInterrupcao', 'NomAgenteRegulado', 'DscFatoGeradorInterrupcao', 'DatFimInterrupcao', 'DatInicioInterrupcao', 'DscSubestacaoDistribuicao', 'DscAlimentadorSubestacao', 'NumConsumidorConjunto', 'NumNivelTensao']
dtype = {'NumOrdemInterrupcao': str, 'NomAgenteRegulado': str, 'DscFatoGeradorInterrupcao': str, 'DatFimInterrupcao': str, 'DatInicioInterrupcao': str, 'DscSubestacaoDistribuicao': str, 'DscAlimentadorSubestacao': str, 'NumConsumidorConjunto': float, 'NumNivelTensao': float}
df = pd.read_csv(file_path, encoding='ISO-8859-1', sep=';', low_memory=False, usecols=columns_needed, dtype=dtype)

# Converter colunas de data para datetime
df['DatFimInterrupcao'] = pd.to_datetime(df['DatFimInterrupcao'], format="%Y-%m-%d %H:%M:%S")
df['DatInicioInterrupcao'] = pd.to_datetime(df['DatInicioInterrupcao'], format="%Y-%m-%d %H:%M:%S")

# Calcular a diferença de tempo em horas
df['DiferencaTempo'] = (df['DatFimInterrupcao'] - df['DatInicioInterrupcao']).dt.total_seconds() / 3600

# Remover linhas com valores ausentes nas colunas selecionadas para a detecção de outliers
df = df.dropna(subset=['DiferencaTempo', 'NumConsumidorConjunto'])

# Função para detecção de outliers usando IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Detectar outliers nas colunas 'DiferencaTempo' e 'NumConsumidorConjunto'
outliers_diferenca_tempo = detect_outliers_iqr(df, 'DiferencaTempo')
outliers_num_consumidor_conjunto = detect_outliers_iqr(df, 'NumConsumidorConjunto')

# Criar boxplot para a coluna 'DiferencaTempo'
plt.figure(figsize=(10, 6))
plt.boxplot(df['DiferencaTempo'], vert=False)
plt.title('Boxplot da Diferença de Tempo')
plt.xlabel('Diferença de Tempo (horas)')
plt.show()
```

**Descrição:** Este boxplot mostra a distribuição da diferença de tempo em horas para as interrupções de energia. Similar ao primeiro boxplot, a caixa central representa o intervalo interquartil, enquanto os "bigodes" se estendem até 1,5 vezes o IQR a partir de Q1 e Q3. Pontos fora desse intervalo são considerados outliers.

**Interpretação:** A maioria dos dados está concentrada em diferenças de tempo menores, abaixo de aproximadamente 100 horas.
Há um número significativo de outliers acima de 100 horas, indicando interrupções de energia que duraram excepcionalmente mais tempo.

```python
# Criar scatter plot para a coluna 'DiferencaTempo' com outliers destacados
plt.figure(figsize=(10, 6))
plt.scatter(df.index, df['DiferencaTempo'], label='Dados')
plt.scatter(outliers_diferenca_tempo.index, outliers_diferenca_tempo['DiferencaTempo'], color='red', label='Outliers')
plt.title('Scatter Plot da Diferença de Tempo')
plt.xlabel('Índice')
plt.ylabel('Diferença de Tempo (horas)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression

# Carregar o arquivo CSV
file_path = 'C:/Users/u11886/Desktop/TCC 1/interrupcoes-energia-eletrica-2024.csv'
columns_needed = ['NumOrdemInterrupcao', 'NomAgenteRegulado', 'DscFatoGeradorInterrupcao', 'DatFimInterrupcao', 'DatInicioInterrupcao', 'DscSubestacaoDistribuicao', 'DscAlimentadorSubestacao', 'NumConsumidorConjunto', 'NumNivelTensao']
dtype = {'NumOrdemInterrupcao': str, 'NomAgenteRegulado': str, 'DscFatoGeradorInterrupcao': str, 'DatFimInterrupcao': str, 'DatInicioInterrupcao': str, 'DscSubestacaoDistribuicao': str, 'DscAlimentadorSubestacao': str, 'NumConsumidorConjunto': float, 'NumNivelTensao': float}
df = pd.read_csv(file_path, encoding='ISO-8859-1', sep=';', low_memory=False, usecols=columns_needed, dtype=dtype)

# Converter colunas de data para datetime
df['DatFimInterrupcao'] = pd.to_datetime(df['DatFimInterrupcao'], format="%Y-%m-%d %H:%M:%S")
df['DatInicioInterrupcao'] = pd.to_datetime(df['DatInicioInterrupcao'], format="%Y-%m-%d %H:%M:%S")

# Calcular a diferença de tempo em horas
df['DiferencaTempo'] = (df['DatFimInterrupcao'] - df['DatInicioInterrupcao']).dt.total_seconds() / 3600

# Remover linhas com valores ausentes nas colunas selecionadas
df = df.dropna(subset=['DiferencaTempo', 'NumConsumidorConjunto'])

# Função para remover os 1% valores mais extremos
def remove_extreme_percentile(data, column, percentile=0.01):
    lower_bound = data[column].quantile(percentile)
    upper_bound = data[column].quantile(1 - percentile)
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)].copy()

# Remover os 1% valores mais extremos na coluna 'DiferencaTempo'
df_no_outliers = remove_extreme_percentile(df, 'DiferencaTempo', percentile=0.01)

# Verificar a forma dos dados após a remoção dos outliers
print(df_no_outliers.shape)

# Criar a variável alvo binária para 'DiferencaTempo'
def create_outlier_target(data, column):
    data = data.copy()  # Faz uma cópia explícita do DataFrame
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data['Outlier'] = ((data[column] < lower_bound) | (data[column] > upper_bound)).astype(int)
    return data

df_no_outliers = create_outlier_target(df_no_outliers, 'DiferencaTempo')

# Verificar a distribuição das classes após a remoção dos outliers
print(df_no_outliers['Outlier'].value_counts())
```

```python
# Selecionar as colunas de interesse para a regressão logística
features = ['NumConsumidorConjunto', 'NumNivelTensao', 'DiferencaTempo']
target = 'Outlier'

X = df_no_outliers[features]
y = df_no_outliers[target]

# Verificar a distribuição das classes após a remoção dos outliers
print("Distribuição das classes após a remoção dos outliers:")
print(y.value_counts())

# Validação cruzada
kf = StratifiedKFold(n_splits=5)
model = LogisticRegression(max_iter=1000)

scores = cross_val_score(model, X, y, cv=kf, scoring='roc_auc')

print(f'AUC-ROC scores for each fold after removing extreme percentiles: {scores}')
print(f'Mean AUC-ROC score after removing extreme percentiles: {np.mean(scores)}')
```

```python
import matplotlib.pyplot as plt

# Resultados da validação cruzada após remoção ajustada de outliers
scores_no_outliers = scores
mean_score_no_outliers = np.mean(scores_no_outliers)

# Plotando o gráfico de barras
folds = ['Fold 1', 'Fold 2', 'Fold 3', 'Fold 4', 'Fold 5']
plt.figure(figsize=(10, 6))
bars = plt.bar(folds, scores_no_outliers, color='skyblue')

# Adicionando a linha da média
plt.axhline(y=mean_score_no_outliers, color='r', linestyle='--', linewidth=2, label=f'Mean AUC-ROC: {mean_score_no_outliers:.2f}')

# Adicionando rótulos de dados
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, f'{yval:.2f}', ha='center', va='bottom')

# Configurações do gráfico
plt.xlabel('Folds')
plt.ylabel('AUC-ROC Score')
plt.title('AUC-ROC Scores for Each Fold with Mean Score after Adjusted Outlier Removal')
plt.legend()
plt.ylim(0.4, 1.05)  # Ajustando o limite do eixo y para melhor visualização
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Carregar o arquivo CSV
file_path = 'C:/Users/u11886/Desktop/TCC 1/interrupcoes-energia-eletrica-2024.csv'
columns_needed = ['NumConsumidorConjunto', 'NumNivelTensao', 'DatFimInterrupcao', 'DatInicioInterrupcao']
df = pd.read_csv(file_path, encoding='ISO-8859-1', sep=';', usecols=columns_needed)

# Pré-processamento
df = df.dropna()  # Remover valores ausentes

# Converter colunas de data para datetime
df['DatFimInterrupcao'] = pd.to_datetime(df['DatFimInterrupcao'], format="%Y-%m-%d %H:%M:%S")
df['DatInicioInterrupcao'] = pd.to_datetime(df['DatInicioInterrupcao'], format="%Y-%m-%d %H:%M:%S")

# Calcular a diferença de tempo em horas
df['DiferencaTempo'] = (df['DatFimInterrupcao'] - df['DatInicioInterrupcao']).dt.total_seconds() / 3600

# Selecionar as colunas de interesse
X = df[['NumConsumidorConjunto', 'NumNivelTensao', 'DiferencaTempo']]

# Normalização dos dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Aplicar PCA
pca = PCA(n_components=2)  # Reduzir para 2 componentes principais
X_pca = pca.fit_transform(X_scaled)

# Adicionar os componentes principais ao dataframe original para visualização
df['PCA1'] = X_pca[:, 0]
df['PCA2'] = X_pca[:, 1]

# Visualizar os componentes principais
plt.figure(figsize=(10, 6))
plt.scatter(df['PCA1'], df['PCA2'], c='blue', alpha=0.5)
plt.title('PCA: Componentes Principais das Interrupções de Energia')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()

# Explicar a variação dos componentes principais
explained_variance = pca.explained_variance_ratio_
print(f'Variância explicada pelo Componente Principal 1: {explained_variance[0]:.2f}')
print(f'Variância explicada pelo Componente Principal 2: {explained_variance[1]:.2f}')
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Carregar o arquivo CSV
file_path = 'C:/Users/u11886/Desktop/TCC 1/interrupcoes-energia-eletrica-2024.csv'
columns_needed = ['NumConsumidorConjunto', 'NumNivelTensao', 'DatInicioInterrupcao', 'DatFimInterrupcao']
df = pd.read_csv(file_path, encoding='ISO-8859-1', sep=';', usecols=columns_needed)

# Converter colunas de data para datetime
df['DatFimInterrupcao'] = pd.to_datetime(df['DatFimInterrupcao'], format="%Y-%m-%d %H:%M:%S")
df['DatInicioInterrupcao'] = pd.to_datetime(df['DatInicioInterrupcao'], format="%Y-%m-%d %H:%M:%S")

# Calcular a diferença de tempo em horas
df['DiferencaTempo'] = (df['DatFimInterrupcao'] - df['DatInicioInterrupcao']).dt.total_seconds() / 3600

# Selecionar as colunas relevantes para o clustering
df = df[['NumConsumidorConjunto', 'NumNivelTensao', 'DiferencaTempo']].dropna()

# Reduzir a amostra dos dados (por exemplo, 1% dos dados)
df_sample = df.sample(frac=0.01, random_state=42)

# Padronizar os dados
X = df_sample.values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Aplicar DBSCAN com parâmetros ajustados
db = DBSCAN(eps=0.5, min_samples=10)
labels = db.fit_predict(X_scaled)

# Adicionar os labels ao DataFrame
df_sample['Cluster'] = labels

# Visualizar os resultados
plt.figure(figsize=(10, 6))
unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

for k, col in zip(unique_labels, colors):
    if k == -1:
        # Cor preta para ruído
        col = [0, 0, 0, 1]

    class_member_mask = (labels == k)
    xy = X_scaled[class_member_mask]

    plt.plot(xy[:, 0], xy[:, 2], 'o', markerfacecolor=tuple(col), markeredgecolor='k', markersize=6)

plt.title('DBSCAN: Clusters de Interrupções de Energia e Outliers (Amostra Reduzida)')
plt.xlabel('Número de Consumidores por Conjunto (Padronizado)')
plt.ylabel('Duração da Interrupção (Padronizado)')
plt.show()
```

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Carregar os dados
file_path = 'C:/Users/u11886/Desktop/TCC 1/interrupcoes-energia-eletrica-2024.csv'
columns_needed = ['NumConsumidorConjunto', 'NumNivelTensao', 'DatFimInterrupcao', 'DatInicioInterrupcao']
df = pd.read_csv(file_path, encoding='ISO-8859-1', sep=';', usecols=columns_needed)

# Pré-processamento
df['DatFimInterrupcao'] = pd.to_datetime(df['DatFimInterrupcao'], format="%Y-%m-%d %H:%M:%S")
df['DatInicioInterrupcao'] = pd.to_datetime(df['DatInicioInterrupcao'], format="%Y-%m-%d %H:%M:%S")
df['DiferencaTempo'] = (df['DatFimInterrupcao'] - df['DatInicioInterrupcao']).dt.total_seconds() / 3600

df = df.dropna(subset=['NumConsumidorConjunto', 'NumNivelTensao', 'DiferencaTempo'])  # Remover valores ausentes

# Amostragem dos dados
df_sample = df.sample(frac=0.01, random_state=42)

# Normalizar os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_sample[['NumConsumidorConjunto', 'NumNivelTensao', 'DiferencaTempo']])

# Aplicar Isolation Forest
iso_forest = IsolationForest(contamination=0.01, random_state=42)
labels = iso_forest.fit_predict(X_scaled)

# Adicionar os labels ao DataFrame
df_sample['Outlier'] = labels

# Visualizar os resultados
plt.figure(figsize=(10, 6))
plt.scatter(df_sample['NumConsumidorConjunto'], df_sample['DiferencaTempo'], c=df_sample['Outlier'], cmap='coolwarm', edgecolor='k')
plt.title('Isolation Forest: Detecção de Outliers nas Interrupções de Energia (Amostra Reduzida)')
plt.xlabel('Número de Consumidores por Conjunto')
plt.ylabel('Duração da Interrupção (horas)')
plt.show()
```

### Comparação dos Métodos de Detecção de Outliers Utilizados

#### 1. **Método do Intervalo Interquartil (IQR)**
   - **Descrição**: Utiliza os quartis para identificar outliers. Valores fora do intervalo [Q1 - 1.5 * IQR, Q3 + 1.5 * IQR] são considerados outliers.
   - **Vantagens**: Simples e intuitivo.
   - **Desvantagens**: Pode não ser eficaz para distribuições não normais ou quando há muitos outliers.
   - **Resultados**: Identificou uma quantidade moderada de outliers baseada na distribuição das durações das interrupções.

#### 2. **Z-Score**
   - **Descrição**: Mede o número de desvios padrão que um ponto de dados está da média. Valores com Z-Score superior a 3 ou inferior a -3 são considerados outliers.
   - **Vantagens**: Fácil de aplicar e interpretar.
   - **Desvantagens**: Assumção de distribuição normal dos dados pode limitar sua eficácia.
   - **Resultados**: Identificou outliers em pontos que estavam a mais de três desvios padrão da média das durações das interrupções.

#### 3. **Regressão Logística**
   - **Descrição**: Modelo estatístico usado para predizer a probabilidade de um evento binário, neste caso, a presença de outliers.
   - **Vantagens**: Permite a inclusão de múltiplas variáveis para a identificação de outliers.
   - **Desvantagens**: Complexidade maior e requer validação adequada.
   - **Resultados**: A análise estatística identificou variáveis que influenciam a ocorrência de outliers e gerou uma matriz de confusão e curva ROC para avaliar a performance do modelo.

#### 4. **K-Means**
   - **Descrição**: Algoritmo de clustering que particiona dados em K clusters. Outliers são pontos que não se encaixam bem em nenhum cluster.
   - **Vantagens**: Bom para detectar outliers em grandes datasets.
   - **Desvantagens**: Requer a escolha do número de clusters e pode ser sensível a outliers.
   - **Resultados**: Segmentou interrupções em clusters, identificando pontos que não se encaixaram bem como possíveis outliers.

#### 5. **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**
   - **Descrição**: Algoritmo de clustering baseado em densidade que encontra áreas de alta densidade separadas por áreas de baixa densidade. Pontos em áreas de baixa densidade são considerados outliers.
   - **Vantagens**: Não requer a definição do número de clusters e é robusto a outliers.
   - **Desvantagens**: Sensível à escolha dos parâmetros epsilon e min_samples.
   - **Resultados**: Identificou clusters de interrupções e pontos isolados como outliers, mostrando uma boa capacidade de detecção em áreas de baixa densidade.

#### 6. **Isolation Forest**
   - **Descrição**: Algoritmo de aprendizado não supervisionado que isola observações ao construir árvores de decisão. Outliers são pontos que são isolados rapidamente.
   - **Vantagens**: Escalável para grandes conjuntos de dados e eficaz na detecção de outliers.
   - **Desvantagens**: Parâmetros de contaminação precisam ser ajustados corretamente.
   - **Resultados**: Detectou outliers com base na facilidade de isolamento de pontos anômalos, visualizados em gráficos que mostram separação clara entre outliers e dados normais.

### Conclusão

Após a análise comparativa dos métodos de detecção de outliers aplicados às interrupções de energia elétrica nas redes de distribuição, observamos que cada método apresentou características distintas em termos de eficiência e adequação aos dados.

- **Isolation Forest** e **DBSCAN** se destacaram na detecção de outliers em grandes conjuntos de dados e em cenários complexos. **Isolation Forest** foi particularmente eficaz em identificar outliers isolados rapidamente, enquanto **DBSCAN** foi robusto na identificação de outliers em áreas de baixa densidade, sem a necessidade de definir o número de clusters.

- **Regressão Logística** forneceu uma análise detalhada das variáveis que influenciam a ocorrência de outliers, permitindo uma compreensão mais profunda dos fatores determinantes. No entanto, sua complexidade e necessidade de validação rigorosa tornam-na menos prática para implementações rápidas.

- **K-Means** foi útil para segmentação e identificação de padrões, mas sua sensibilidade à escolha do número de clusters e a presença de outliers dentro dos clusters limitou sua eficácia para detecção pura de outliers.

- **IQR** e **Z-Score** são métodos mais simples e rápidos de aplicar, sendo eficazes para análises preliminares. No entanto, sua aplicação é limitada em distribuições não normais ou em datasets com grande quantidade de outliers.

Dado o contexto do trabalho, onde temos um dataset grande e complexo, os métodos Isolation Forest e DBSCAN foram os mais eficazes. Isolation Forest foi particularmente eficiente em detectar outliers de forma rápida e escalável, enquanto DBSCAN mostrou-se robusto em encontrar outliers em áreas de baixa densidade. Regressão Logística proporcionou uma compreensão profunda dos fatores que influenciam a presença de outliers, apesar de sua maior complexidade. Para análises preliminares, IQR e Z-Score se mostraram úteis, mas com limitações consideráveis. Por fim, K-Means ajudou na segmentação inicial dos dados, mas com restrições na detecção de outliers puramente.
